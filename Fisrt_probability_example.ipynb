1.  Ayudame a hacer un ejercicio en python de probabilidad, aplicando el teorema de bayes
2. Tengo dos cadas roja(R) y azul (B) y tengo dos tipos de frutas manzana(a) y naranja(o)
que estan distrubuidas en las cajas, asi como muestra la imagen.
X es la variable aleatoria de la caja  
Y es la variable alaeatoria de la fruta. Necesito calcular todas las probabilidades condicionales y las probabilidades marginles de todos los posibles eventos




```python
def main():
    # Probabilidades de las cajas (P(X))
    P_R = 0.4  # Probabilidad de elegir la Caja Roja
    P_B = 0.6  # Probabilidad de elegir la Caja Azul


    # Probabilidades condicionales de las frutas en cada caja (P(Y|X))
    P_A_given_R = 2/8  # Probabilidad de Manzana dado que es Caja Roja
    P_O_given_R = 6/8  # Probabilidad de Naranja dado que es Caja Roja
    P_A_given_B = 3/4  # Probabilidad de Manzana dado que es Caja Azul
    P_O_given_B = 1/4  # Probabilidad de Naranja dado que es Caja Azul

    # ===== Cálculo de probabilidades marginales de Y (P(Y)) =====
    # P(Y=A) = P(A|R)*P(R) + P(A|B)*P(B)
    P_A = P_A_given_R * P_R + P_A_given_B * P_B
    # P(Y=O) = P(O|R)*P(R) + P(O|B)*P(B)
    P_O = P_O_given_R * P_R + P_O_given_B * P_B

    # ===== Cálculo de probabilidades condicionales inversas (P(X|Y)) =====
    # Usamos el Teorema de Bayes:
    # P(R|A) = P(A|R) * P(R) / P(A)
    P_R_given_A = (P_A_given_R * P_R) / P_A
    P_B_given_A = (P_A_given_B * P_B) / P_A
    P_R_given_O = (P_O_given_R * P_R) / P_O
    P_B_given_O = (P_O_given_B * P_B) / P_O

    # ===== Resultados =====
    print("\n=== Probabilidades Marginales ===")
    print(f"P(Y=A): {P_A:.3f}")  # Probabilidad total de Manzana
    print(f"P(Y=O): {P_O:.3f}")  # Probabilidad total de Naranja

    print("\n=== Probabilidades Condicionales Inversas ===")
    print(f"P(X=R|Y=A): {P_R_given_A:.3f}")  # Dado Manzana, prob. de Caja Roja
    print(f"P(X=B|Y=A): {P_B_given_A:.3f}")  # Dado Manzana, prob. de Caja Azul
    print(f"P(X=R|Y=O): {P_R_given_O:.3f}")  # Dado Naranja, prob. de Caja Roja
    print(f"P(X=B|Y=O): {P_B_given_O:.3f}")  # Dado Naranja, prob. de Caja Azul

if __name__ == "__main__":
    main()
```

```python
def calcular_probabilidades(caja_roja, caja_azul, prob_caja_roja):
    """
    Calcula todas las probabilidades marginales y condicionales del problema.

    Args:
        caja_roja (dict): Diccionario con las cantidades de manzanas y naranjas en la caja roja.
                          Ejemplo: {"manzanas": 2, "naranjas": 6}
        caja_azul (dict): Diccionario con las cantidades de manzanas y naranjas en la caja azul.
                          Ejemplo: {"manzanas": 3, "naranjas": 1}
        prob_caja_roja (float): Probabilidad de elegir la caja roja (ej: 0.4).
                               La probabilidad de la caja azul se calcula como 1 - prob_caja_roja.
    """
    # Probabilidades de las cajas
    P_R = prob_caja_roja
    P_B = 1 - P_R

    # Calcular probabilidades condicionales P(Y|X) para cada caja
    total_roja = caja_roja["manzanas"] + caja_roja["naranjas"]
    P_A_given_R = caja_roja["manzanas"] / total_roja
    P_O_given_R = caja_roja["naranjas"] / total_roja

    total_azul = caja_azul["manzanas"] + caja_azul["naranjas"]
    P_A_given_B = caja_azul["manzanas"] / total_azul
    P_O_given_B = caja_azul["naranjas"] / total_azul

    # Probabilidades marginales P(Y)
    P_A = P_A_given_R * P_R + P_A_given_B * P_B
    P_O = P_O_given_R * P_R + P_O_given_B * P_B

    # Probabilidades condicionales inversas P(X|Y) usando Bayes
    P_R_given_A = (P_A_given_R * P_R) / P_A if P_A != 0 else 0
    P_B_given_A = (P_A_given_B * P_B) / P_A if P_A != 0 else 0
    P_R_given_O = (P_O_given_R * P_R) / P_O if P_O != 0 else 0
    P_B_given_O = (P_O_given_B * P_B) / P_O if P_O != 0 else 0

    # Resultados
    print("\n=== Probabilidades Marginales ===")
    print(f"P(Y=A): {P_A:.3f}")  # Probabilidad total de Manzana
    print(f"P(Y=O): {P_O:.3f}")  # Probabilidad total de Naranja

    print("\n=== Probabilidades Condicionales Inversas ===")
    print(f"P(X=R|Y=A): {P_R_given_A:.3f}")  # Dado Manzana, prob. de Caja Roja
    print(f"P(X=B|Y=A): {P_B_given_A:.3f}")  # Dado Manzana, prob. de Caja Azul
    print(f"P(X=R|Y=O): {P_R_given_O:.3f}")  # Dado Naranja, prob. de Caja Roja
    print(f"P(X=B|Y=O): {P_B_given_O:.3f}")  # Dado Naranja, prob. de Caja Azul

def main():
    # Ejemplo de uso:
    caja_roja = {"manzanas": 2, "naranjas": 6}  # 2 manzanas y 6 naranjas en caja roja
    caja_azul = {"manzanas": 3, "naranjas": 1}  # 3 manzanas y 1 naranja en caja azul
    prob_caja_roja = 0.4  # Probabilidad de elegir la caja roja (40%)

    calcular_probabilidades(caja_roja, caja_azul, prob_caja_roja)

if __name__ == "__main__":
    main()
```
